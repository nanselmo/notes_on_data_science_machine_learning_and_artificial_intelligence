```python
from os import listdir
```

```python
filenames = []

for file in listdir("."):
    if file.endswith(".md"):
        filenames.append(file)
```

```python
print(str('![png]({filename}/images/' + slug + '/'))
```

```python
for file in filenames:
    with open(file, 'r+') as f:
        content = f.read()
        slug = str(file).split('.')[0]
        content =  content.replace('![png](', '![png]({filename}/images/' + slug + '/')
        
        f.seek(0)
        f.truncate()
        
        f.write(content)
```
