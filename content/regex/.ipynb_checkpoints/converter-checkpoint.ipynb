```python
from os import listdir
```

```python
filenames = []

for file in listdir("."):
    if file.endswith(".md"):
        filenames.append(file)
```

```python
filenames
```

```python
with open(filenames[0], 'r+') as f:
    content = f.read()
    header = 'Title: TITLETITLE\nSlug: SLUGSLUG\nSummary: TITLETITLE\nDate: 2016-12-01 12:00\nCategory: Regex\nTags: Basics\nAuthors: Chris Albon\n'
    title = str(content.split('\n')[1].lstrip('# '))
    slug = str(filenames[0]).split('.')[0]
    header = header.replace('TITLETITLE', title)
    header = header.replace('SLUGSLUG', slug)
    
    content = content.replace(str(content.split('\n')[1]), '')
    
    f.seek(0, 0)
    f.write(header + content)
```
